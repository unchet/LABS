```
det = Det[A - IdentityMatrix[3]*l];
det // TraditionalForm
If [CharacteristicPolynomial[A, l] == det, "Характеристический многочлен совпадает с найденным вручную", "Что-то пошло не так"]
```

```
slns = N[Reduce[det == 0, l]];
eigenvalues = {};
For[i = 1, i < 4, i++,
    eigenvalues = Append[eigenvalues, slns[[i,2]]];
];
eigenvalues // TraditionalForm
Sort[N[Eigenvalues[A]]] // TraditionalForm
If[Sort[Eigenvalues[A]] == eigenvalues, "Собственные значения совпадают с найденными вручную", "Что-то пошло не так"]
```

```
B1 = A - IdentityMatrix[3] * eigenvalues[[1]];
B2 = A - IdentityMatrix[3] * eigenvalues[[2]];
B3 = A - IdentityMatrix[3] * eigenvalues[[3]];
MatrixForm[B1] // TraditionalForm
MatrixForm[B2] // TraditionalForm
MatrixForm[B3] // TraditionalForm
```

Находим собственные векторы

```
xyz := {x, y, z};
v1 := B1.xyz;
res1 = Solve[v1 == {0, 0, 0}];
res1 = Prepend[{y,z}/.res1[[1]]/.x->1, 1];
res1 // TraditionalForm
v2 := B2.xyz;
res2 = Solve[v2 == {0, 0, 0}];
res2 = Prepend[{y,z}/.res2[[1]]/.x->1, 1];
res2 // TraditionalForm
v3 := B3.xyz;
res3 = Solve[v3 == {0, 0, 0}];
res3 = Prepend[{y,z}/.res3[[1]]/.x->1, 1];
res3 // TraditionalForm
```

Сравниваем с собственными векторами, найденными wolfram

Проверяем, что мы действительно нашли собственные векторы. Составляем матрицу S из нормированных векторов 

```
diag := {
    {eigenvalues[[1]], 0, 0},
    {0, eigenvalues[[2]], 0},
    {0, 0, eigenvalues[[3]]}
};
vn1 := Normalize[res1];
vn2 := Normalize[res2];
vn3 := Normalize[res3];
S := {vn1, vn2, vn3};
MatrixForm[S] // TraditionalForm
check := Transpose[S].diag.S;
MatrixForm[check] // TraditionalForm
```
