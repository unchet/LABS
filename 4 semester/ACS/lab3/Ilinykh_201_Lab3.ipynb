### Задание 2.  Падение тела, брошенного под углом к горизонту

```python
var("y1 y2 y3 y4 t")

# начальные условия
k = 0.01
g = 9.81
tmax = 1.41
alpha = pi/4
v0 = 10.0
y0 = [0, 0, v0*cos(alpha), 0, v0*sin(alpha)]

# правые части ОДУ
ode = [y2, -k*y2*sqrt(y2^2 + y4^2), y4, -k*y4*sqrt(y2^2 + y4^2) - g]

# метод Рунге – Кутта
solutions = desolve_system_rk4(des = ode, vars = [y1, y2, y3, y4], ics = y0, ivar = t, step = 0.01, end_points = tmax)

lines = plot([])

for i in range(len(solutions) - 1):
    lines += line([[solutions[i][1], solutions[i][3]], [solutions[i + 1][1], solutions[i + 1][3]]])

show(lines)

forget("y1 y2 y3 y4 t")
```
