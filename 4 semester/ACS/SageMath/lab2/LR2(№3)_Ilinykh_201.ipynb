```python
plot_range = 10.0
u(x, y, z) = 11 * x ^ 2 - 2 * x * y - 2 * x * z + 2 * y * z + 9 * z ^ 2 - 4 * x + y + z
show(u(x, y, z) == 0)
```

```python
a11 = 11
a22 = 0
a33 = 9

a12 = -2 / 2
a13 = -2 / 2
a23 = 2 / 2

a0 = 0
a1 = -4 / 2
a2 = 1 / 2
a3 = 1 / 2

matrA = matrix([
    [a11, a12, a13],
    [a12, a22, a23],
    [a13, a23, a33]
])

vecA = vector([a1, a2, a3])

show(LatexExpr("A = " + latex(matrA)))
show(LatexExpr("a = " + latex(vecA)))
```

```python
var("l")
matrAm = matrA - l * identity_matrix(3)
detA = det(matrAm).simplify_full()
show(LatexExpr(r'\det(A - l * E) = ' + latex(detA)))
```

```python
solutions = solve(detA == 0, l)
eigenvalues = []
for i in range(len(solutions)):
    eigenvalues.append(solutions[i].rhs().real().n(digits = 8))
    show(LatexExpr("l_{" + str(i) + "} = " + str(eigenvalues[i])))
```

```python
print("Сравнение с методом eigenvalues:")
show(matrA.eigenvalues())
```

```python
eigenvectors = []
var("x1 x2 x3")
for i in range(len(eigenvalues)):
    matrB = matrA - eigenvalues[i] * identity_matrix(3)
    system = []
    for j in range(3):
        system.append(matrB[j][0] * x1 + matrB[j][1] * x2 + matrB[j][2] * x3 == 0)
    
    system[1] = system[1] - system[0] * matrB[1][0] / matrB[0][0]
    system[2] = system[2] - system[0] * matrB[2][0] / matrB[0][0]
    system[2] = system[2] - system[1] * (system[2].lhs() / system[1].lhs())
    
    if (system[2].lhs() == 0 and system[2].rhs() == 0):
        system[2] = (x3 == 1)
    else:
        print("Ранг матрицы равен трём. Что-то пошло не так")
        
    eigenvec = vector([k.rhs().n(digits = 8) for k in solve(system, x1, x2, x3)[0]])
    ans = matrB * eigenvec
    show(LatexExpr(r'\lambda_{' + str(i) + '} = ' + str(eigenvalues[i].n(digits = 8)) + ', s_{' + str(i) + '} = ' + str(eigenvec.n(digits = 8))))
    eigenvectors.append(eigenvec.n(digits = 8))
```

```python
print("Сравнение с методом eigenvectors_right:")
show(matrA.eigenvectors_right())
```

```python
matrS = matrix(eigenvectors)
print("Матрица перехода:")
show(LatexExpr("S = " + latex(matrS.n(digits = 8))))

for i in range(len(eigenvectors)):
    eigenvectors[i] = eigenvectors[i] / sqrt(eigenvectors[i].dot_product(eigenvectors[i]))
matrSn = matrix(eigenvectors)
print("Нормированная матрица перехода:")
show(LatexExpr("S_n = " + latex(matrSn.n(digits = 8))))
```

```python
matrDiag = matrix([
    [eigenvalues[0], 0, 0],
    [0, eigenvalues[1], 0],
    [0, 0, eigenvalues[2]]
])
matrCheck = matrSn.T * matrDiag * matrSn
show(matrCheck)
print("Видим, что эта матрица соответствует изначальной, значит все вычисления были верными")
```

```python
vecA1 = matrSn.T * vecA
show(vecA1)
```

```python
v(x, y, z) = 0
variables = [x, y, z]
for i in range(len(variables)):
    v += matrDiag[i][i] * variables[i] ^ 2 + variables[i] * vecA1[i]
show(v(x, y, z) == 0)
```

```python
coef = []
for i in range(len(vecA1)):
    coef.append((vecA1[i] / (2 * matrDiag[i][i])) ** 2)
    if (matrDiag[i][i] * vecA1[i] < 0):
        coef[i] = -coef[i]

v(x, y, z) = 0
for i in range(3):
    v += matrDiag[i][i] * (variables[i] + coef[i]) ^ 2 - matrDiag[i][i] * abs(coef[i])

show(v(x, y, z) == 0)
```
